```python
#download dataset

from icrawler.builtin import GoogleImageCrawler

# List of heavy machine parts
categories = [
    "Excavator Bucket",
    "Bulldozer Blade",
    "Crane Hook",
    "Loader Tire",
    "Hydraulic Cylinder",
    "Heavy Machinery Engine",
    "Gearbox Heavy Machinery",
    "Axle Heavy Equipment",
    "Undercarriage Bulldozer",
    "Boom Arm Excavator"
]

for category in categories:
    crawler = GoogleImageCrawler(storage={'root_dir': f'heavy_machinery_parts_dataset/{category.replace(" ", "_").lower()}'})
    crawler.crawl(keyword=category, max_num=50)  # ~50 images per part
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

# Image settings
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
DATASET_DIR = "heavy_machinery_parts_dataset"

# Preprocess images
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_data = datagen.flow_from_directory(
    DATASET_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    subset='training'
)

val_data = datagen.flow_from_directory(
    DATASET_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    subset='validation'
)
```
